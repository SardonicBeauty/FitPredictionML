```python
body_map = {"hourglass": 1, "athletic": 2, "petite": 3, "pear": 4, "straight & narrow":5, "full bust":6, "apple":7}
height_map = {'5. 40':1, '5. 60': 2, '5. 50': 3, '5. 70': 4, '5. 30': 5, '5. 20': 6, '5. 80': 7, '5. 90': 8,
               '5. 10':9, '5. 00': 10, '5. 11': 11, '6. 00': 12, '4. 11': 13, '6. 10': 14, '4. 10': 15, '6. 20': 16, '4. 90':54 }
category_map = {'dress':1 , 'gown':2 , 'sheath':3 , 'shift':4 , 'jumpsuit':5 , 'jacket':6 , 'sweatshirt':7, 'romper':8}
```

```python
df['body_type'] = df['body_type'].map(body_map) #can use replace also .replace()
df['height'] = df['height'].map(height_map)
df['category'] = df['category'].map(category_map)
```
